```python
import pandas as pd
```

```python
dataset = pd.read_csv("tweets.csv", encoding = "ISO-8859-1")
```

```python
dataset.head(5)
```

```python
def gen_freq(text):
    word_list = []
    
    for tw_words in text.split():
        word_list.extend(tw_words)

    word_freq = pd.Series(word_list).value_counts()

    return word_freq
    
```

```python
word_freq = gen_freq(dataset.text.str)
word_freq
```
